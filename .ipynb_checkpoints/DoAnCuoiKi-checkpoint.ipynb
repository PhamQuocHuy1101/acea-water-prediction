```python
# convert string to datetime
def to_date(df):
    df.Date = pd.to_datetime(df.Date, format='%d/%m/%Y')
```

```python
# Check date missing
delta = df.Date - df.Date.shift(1)
delta.sum(), delta.count() # duration 1 days in all data
```

```python
def plot_null_value(df, columns):
    fig, axs = plt.subplots(nrows=len(columns), ncols=1, figsize=(10, 5 * len(columns)))
    if len(columns) == 1:
        axs = [axs]
    for ax, col in zip(axs, columns):
        null_value = df[col].fillna(0)
        replace = df[col].fillna(np.inf)

        sns.lineplot(x=df.Date, y=null_value, ax=ax, color='darkorange', label='miss value')
        sns.lineplot(x=df.Date, y=replace, ax=ax, color='dodgerblue', label='fill value')

        ax.set_xlim([df.Date.min(), df.Date.max()])
```

```python
df = pd.read_csv('./data/Aquifer_Petrignano.csv') # Good 8
```

```python
df.info()
```

```python
to_date(df)
plot_null_value(df, df.columns)
```

```python
import datetime
```

```python
# check null value
df.Date[df.Rainfall_Bastia_Umbra.isnull()].max(), df.Date[df.Temperature_Bastia_Umbra.isnull()].max(), df.Date[df.Temperature_Petrignano.isnull()].max(), df.Date[df.Hydrometry_Fiume_Chiascio_Petrignano.isnull()].max()
```

```python
# Remove value from 31/12/2008
df = df[df.Date > datetime.datetime(2008, 12, 31)]
df.index = np.arange(len(df))
df.head()
```

```python
sns.heatmap(df.corr())
```

```python
plot_null_value(df, df.columns)
```

```python
df.info()
```

```python
df.Volume_C10_Petrignano.interpolate(inplace=True)
df.info()
```

```python
delta = df.Depth_to_Groundwater_P24 - df.Depth_to_Groundwater_P25
delta.mean(), delta.std()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df.Date, y=df.Depth_to_Groundwater_P24, color='red')
sns.lineplot(x=df.Date, y=df.Depth_to_Groundwater_P25, color='blue')
plt.show()
```

```python
sns.distplot(delta)
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df.Date, y=df.Depth_to_Groundwater_P24 - delta.mean(), color='red')
sns.lineplot(x=df.Date, y=df.Depth_to_Groundwater_P25, color='blue')
plt.show()
```

```python
((df.Depth_to_Groundwater_P24 - delta.mean()) - df.Depth_to_Groundwater_P25).mean(), ((df.Depth_to_Groundwater_P25 + delta.mean()) - df.Depth_to_Groundwater_P24).mean()
```

```python
p24_null_index = df.Depth_to_Groundwater_P24[df.Depth_to_Groundwater_P24.isnull()].index
p25_null_index = df.Depth_to_Groundwater_P25[df.Depth_to_Groundwater_P25.isnull()].index
```

```python
p24_fill = df.Depth_to_Groundwater_P24.copy()
p24_fill[p24_null_index] = df.Depth_to_Groundwater_P25[p24_null_index] + delta.mean()
p24_fill
```

```python
p25_fill = df.Depth_to_Groundwater_P25.copy()
p25_fill[p25_null_index] = df.Depth_to_Groundwater_P24[p25_null_index] - delta.std()
p25_fill
```

```python
df.Depth_to_Groundwater_P24 = p24_fill
df.Depth_to_Groundwater_P25 = p25_fill

plot_null_value(df, ['Depth_to_Groundwater_P24', 'Depth_to_Groundwater_P25'])
```
